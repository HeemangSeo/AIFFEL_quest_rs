```python
import cv2
import dlib
import numpy as np
import os
import matplotlib.pyplot as plt

# =================================================
# 경로 설정
# =================================================
BASE_PATH = "/home/jovyan/work/AIFFEL_quest_rs/Exploration/Ex01/camera_sticker"

FACE_IMG_PATH = f"{BASE_PATH}/images/face.png"
WHISKER_IMG_PATH = f"{BASE_PATH}/images/cat-whiskers.png"
PREDICTOR_PATH = f"{BASE_PATH}/models/shape_predictor_68_face_landmarks.dat"

# =================================================
# 이미지 로드
# =================================================
img = cv2.imread(FACE_IMG_PATH)
if img is None:
    raise FileNotFoundError("face.png 로드 실패")

img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

whisker = cv2.imread(WHISKER_IMG_PATH, cv2.IMREAD_UNCHANGED)
if whisker is None:
    raise FileNotFoundError("cat-whiskers.png 로드 실패")

# =================================================
# 얼굴 검출 & 랜드마크
# =================================================
detector = dlib.get_frontal_face_detector()
predictor = dlib.shape_predictor(PREDICTOR_PATH)

faces = detector(img_rgb, 1)
if len(faces) == 0:
    raise ValueError("얼굴을 찾지 못했습니다.")

face = faces[0]
landmarks = predictor(img_rgb, face)

# =================================================
# 수염 크기 설정
# =================================================
face_width = face.right() - face.left()
whisker_width = int(face_width * 0.9)

h, w, _ = whisker.shape
whisker_height = int(h * (whisker_width / w))

whisker_resized = cv2.resize(whisker, (whisker_width, whisker_height))

# =================================================
# 수염 색을 검정으로 변경
# =================================================
whisker_black = whisker_resized.copy()
whisker_black[:, :, 0] = 0
whisker_black[:, :, 1] = 0
whisker_black[:, :, 2] = 0

# =================================================
# 입가 기준 위치 계산
# =================================================
mouth_left = landmarks.part(48)
mouth_right = landmarks.part(54)
mouth_center = landmarks.part(62)
nose = landmarks.part(30)

center_x = (mouth_left.x + mouth_right.x) // 2
center_y = int((nose.y + mouth_center.y) / 2)

x1 = int(center_x - whisker_width / 2)
y1 = int(center_y - whisker_height / 2)
x2 = x1 + whisker_width
y2 = y1 + whisker_height

# =================================================
# 이미지 경계 처리
# =================================================
img_h, img_w, _ = img.shape
x1 = max(0, x1)
y1 = max(0, y1)
x2 = min(img_w, x2)
y2 = min(img_h, y2)

whisker_black = whisker_black[0:y2-y1, 0:x2-x1]

# =================================================
# 알파 블렌딩
# =================================================
overlay = img.copy()
alpha = whisker_black[:, :, 3] / 255.0

for c in range(3):
    overlay[y1:y2, x1:x2, c] = (
        alpha * whisker_black[:, :, c] +
        (1 - alpha) * overlay[y1:y2, x1:x2, c]
    )

# =================================================
# 결과 출력
# =================================================
result = cv2.cvtColor(overlay, cv2.COLOR_BGR2RGB)

plt.figure(figsize=(6, 6))
plt.imshow(result)
plt.axis("off")
plt.show()
```
