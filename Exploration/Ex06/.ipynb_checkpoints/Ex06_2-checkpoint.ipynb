### 2. Data Loader (IMDB 스타일)

```python
from konlpy.tag import Okt

tokenizer = Okt()

stopwords = ['의','가','이','은','들','는','좀','잘','걍','과','도',
             '를','으로','자','에','와','한','하다']

def load_data(train_df, test_df, num_words=10000):

    train_df = train_df.drop_duplicates(subset=['document']).dropna()
    test_df  = test_df.drop_duplicates(subset=['document']).dropna()

    X_train = [[w for w in tokenizer.morphs(s) if w not in stopwords]
               for s in train_df['document']]
    X_test  = [[w for w in tokenizer.morphs(s) if w not in stopwords]
               for s in test_df['document']]

    words = np.concatenate(X_train)
    counter = Counter(words).most_common(num_words - 4)

    vocab = ['<PAD>', '<BOS>', '<UNK>', '<UNUSED>'] + [w for w, _ in counter]
    word_to_index = {w: i for i, w in enumerate(vocab)}

    BOS = word_to_index['<BOS>']
    UNK = word_to_index['<UNK>']

    def encode(tokens):
        return [BOS] + [word_to_index.get(w, UNK) for w in tokens]

    X_train = [encode(x) for x in X_train]
    X_test  = [encode(x) for x in X_test]

    y_train = np.array(train_df['label'])
    y_test  = np.array(test_df['label'])

    return X_train, y_train, X_test, y_test, word_to_index
```

```python
X_train, y_train, X_test, y_test, word_to_index = load_data(train_data, test_data)
```

```python
index_to_word = {index:word for word, index in word_to_index.items()}
```

```python
# 문장 1개를 활용할 딕셔너리와 함께 주면, 단어 인덱스 리스트 벡터로 변환해 주는 함수입니다.
# 단, 모든 문장은 <BOS>로 시작하는 것으로 합니다.
def get_encoded_sentence(sentence, word_to_index):
    return [word_to_index['<BOS>']]+[word_to_index[word] if word in word_to_index else word_to_index['<UNK>'] for word in sentence.split()]

# 여러 개의 문장 리스트를 한꺼번에 단어 인덱스 리스트 벡터로 encode해 주는 함수입니다.
def get_encoded_sentences(sentences, word_to_index):
    return [get_encoded_sentence(sentence, word_to_index) for sentence in sentences]

# 숫자 벡터로 encode된 문장을 원래대로 decode하는 함수입니다.
def get_decoded_sentence(encoded_sentence, index_to_word):
    return ' '.join(index_to_word[index] if index in index_to_word else '<UNK>' for index in encoded_sentence[1:])  #[1:]를 통해 <BOS>를 제외

# 여러 개의 숫자 벡터로 encode된 문장을 한꺼번에 원래대로 decode하는 함수입니다.
def get_decoded_sentences(encoded_sentences, index_to_word):
    return [get_decoded_sentence(encoded_sentence, index_to_word) for encoded_sentence in encoded_sentences]
```

### 3. 문장 길이 분석 & Padding

```python
lengths = [len(x) for x in X_train]

plt.hist(lengths, bins=50)
plt.title("Sentence Length Distribution")
plt.show()
```

```python
MAX_LEN = 50
def pad_sequences(seqs, max_len):
    return pad_sequence(
        [torch.tensor(s[:max_len]) for s in seqs],
        batch_first=True,
        padding_value=word_to_index['<PAD>']
    )

X_train_pad = pad_sequences(X_train, MAX_LEN)
X_test_pad  = pad_sequences(X_test, MAX_LEN)

y_train = torch.tensor(y_train)
y_test  = torch.tensor(y_test)
```

### 4. Train / Validation 분리

```python
class NSMCDataset(Dataset):
    def __init__(self, X, y):
        self.X = X
        self.y = y

    def __len__(self):
        return len(self.y)

    def __getitem__(self, idx):
        return self.X[idx], self.y[idx]
```

```python
train_loader = DataLoader(NSMCDataset(X_train_pad, y_train),
                          batch_size=64, shuffle=True)
test_loader  = DataLoader(NSMCDataset(X_test_pad, y_test),
                          batch_size=64)
```

### 5. LSTM 모델 정의

```python
class LSTMClassifier(nn.Module):
    def __init__(self, vocab_size, embed_dim=100, hidden_dim=128):
        super().__init__()
        self.embedding = nn.Embedding(vocab_size, embed_dim, padding_idx=0)
        self.lstm = nn.LSTM(embed_dim, hidden_dim, batch_first=True)
        self.fc = nn.Linear(hidden_dim, 1)

    def forward(self, x):
        x = self.embedding(x)
        _, (h, _) = self.lstm(x)
        return self.fc(h[-1]).squeeze()
```

```python
model = LSTMClassifier(len(word_to_index))
```

### 6. Word2Vec 임베딩 로드 & 적용

```python
from gensim.models import Word2Vec

word2vec_path = os.path.join(DATA_PATH, "word2vec_ko.model")
word_vectors = Word2Vec.load(word2vec_path).wv
```

```python
embedding_dim = word_vectors.vector_size
vocab_size = len(word_to_index)

embedding_matrix = np.zeros((vocab_size, embedding_dim))

for word, idx in word_to_index.items():
    if word in word_vectors:
        embedding_matrix[idx] = word_vectors[word]
```

```python
model.embedding = nn.Embedding.from_pretrained(
    torch.FloatTensor(embedding_matrix),
    padding_idx=word_to_index['<PAD>'],
    freeze=False
)
```

### 7. 모델 학습

```python
criterion = nn.BCEWithLogitsLoss()
optimizer = torch.optim.Adam(model.parameters(), lr=1e-3)
```

```python
def train_epoch(model, loader):
    model.train()
    total_loss, correct = 0, 0

    for x, y in loader:
        optimizer.zero_grad()
        logits = model(x)
        loss = criterion(logits, y.float())
        loss.backward()
        optimizer.step()

        total_loss += loss.item()
        correct += ((torch.sigmoid(logits) > 0.5) == y).sum().item()

    return total_loss / len(loader), correct / len(loader.dataset)
```

```python
def evaluate(model, loader):
    model.eval()
    correct = 0

    with torch.no_grad():
        for x, y in loader:
            logits = model(x)
            correct += ((torch.sigmoid(logits) > 0.5) == y).sum().item()

    return correct / len(loader.dataset)
```

```python
train_accs = []

for epoch in range(5):
    loss, acc = train_epoch(model, train_loader)
    val_acc = evaluate(model, test_loader)
    train_accs.append(acc)

    print(f"[{epoch+1}] loss={loss:.4f}, train_acc={acc:.4f}, val_acc={val_acc:.4f}")
```

### 8. Loss / Accuracy 시각화

```python
plt.plot(train_accs)
plt.title("Training Accuracy")
plt.xlabel("Epoch")
plt.ylabel("Accuracy")
plt.show()
```

### 회고

이번 네이버 영화 리뷰(NSMC) 감성분석 프로젝트를 통해 텍스트 데이터 전처리부터 모델 학습까지의 전체 파이프라인를 직접 구현해볼 수 있었다. IMDB 데이터셋처럼 이미 가공된 데이터가 아니라, **가공되지 않은 원시 텍스트 데이터(raw text)**를 다루면서 자연어 처리에서 전처리가 얼마나 중요한 단계인지 체감할 수 있었다.

가장 먼저 어려움을 느낀 부분은 한국어 형태소 분석 환경 설정이었다. KoNLPy, MeCab, Java(JVM) 설정 과정에서 다양한 에러가 발생했고, 특히 MeCab 사전 경로 문제와 JVMNotFoundException으로 인해 많은 시간을 디버깅에 사용했다. 이 과정을 통해 코드 오류뿐 아니라 실행 환경과 라이브러리 의존성 문제를 해결하는 능력 또한 중요한 개발 역량이라는 점을 깨달았다. 최종적으로는 안정적인 실행을 위해 Okt 형태소 분석기를 사용하여 전처리를 진행하였다.

데이터 로더를 직접 구현하면서는

중복 데이터 제거

결측치 처리

토큰화 및 불용어 제거

단어 사전(word_to_index) 구성

문장을 인덱스 시퀀스로 변환

과 같은 과정을 하나하나 설계해야 했고, 이를 통해 모델 입력 데이터가 만들어지는 전 과정을 명확히 이해할 수 있었다. 또한 문장 길이 분포를 분석하고 padding을 적용하면서, 시퀀스 모델에서 입력 길이 설정이 모델 성능과 학습 안정성에 영향을 준다는 점도 배울 수 있었다.

모델 학습 단계에서는 LSTM 기반 모델을 사용하여 감성 분류를 수행했고, 이후 한국어 Word2Vec 임베딩을 적용하여 사전 학습된 임베딩이 성능 개선에 도움을 줄 수 있음을 확인했다. 특히 임베딩 차원 불일치나 인덱스 범위 오류(IndexError)를 해결하는 과정에서, 모델 구조와 데이터 차원의 일관성 유지가 매우 중요하다는 점을 명확히 인식하게 되었다.

이번 프로젝트를 통해 단순히 모델을 사용하는 것을 넘어,

“텍스트 데이터가 어떻게 숫자로 변환되어 모델에 입력되고,
그 결과가 어떻게 학습으로 이어지는지”

를 전체 흐름 관점에서 이해할 수 있었다는 점이 가장 큰 수확이었다. 향후에는 CNN, GRU 등 다양한 모델 구조를 추가로 실험해보고, 임베딩 시각화나 하이퍼파라미터 튜닝을 통해 성능을 더욱 개선해보고 싶다.
