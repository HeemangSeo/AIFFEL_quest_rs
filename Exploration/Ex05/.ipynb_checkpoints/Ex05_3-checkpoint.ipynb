### ÌòïÌÉúÏÜå Î∂ÑÏÑù (Î™ÖÏÇ¨)

```python
from konlpy.tag import Okt
okt = Okt()

def extract_nouns(text):
    return okt.nouns(text)
```

```python
genre_tokens = {}

for genre, docs in genre_synopsis.items():
    genre_tokens[genre] = [
        extract_nouns(doc)
        for doc in docs
        if len(doc) > 0
    ]
```

### Word2Vec ÌïôÏäµ

```python
sentences = [
    tokens
    for docs in genre_tokens.values()
    for tokens in docs
    if len(tokens) >= 2
]
```

```python
from gensim.models import Word2Vec

w2v_model = Word2Vec(
    sentences=sentences,
    vector_size=100,
    window=5,
    min_count=5,
    workers=4,
    sg=1
)
```

### Embedding ÏùòÎØ∏ Í≤ÄÏ¶ù

### TF-IDFÎ°ú Ïû•Î•¥ ÎåÄÌëú Îã®Ïñ¥ Ï∂îÏ∂ú

```python
from sklearn.feature_extraction.text import TfidfVectorizer
import numpy as np

def get_genre_keywords(genre, top_k=15):
    docs = [' '.join(tokens) for tokens in genre_tokens[genre]]

    vectorizer = TfidfVectorizer()
    tfidf = vectorizer.fit_transform(docs)

    scores = tfidf.mean(axis=0).A1
    words = vectorizer.get_feature_names_out()

    word_scores = sorted(
        zip(words, scores),
        key=lambda x: x[1],
        reverse=True
    )

    keywords = [
        w for w, _ in word_scores
        if w in w2v_model.wv
    ][:top_k]

    return keywords
```

```python
genre_keywords = {
    genre: get_genre_keywords(genre)
    for genre in genre_tokens.keys()
}
```

### Ï§ëÎ≥µ Îã®Ïñ¥ Ï†úÍ±∞

```python
from sklearn.feature_extraction.text import TfidfVectorizer
import numpy as np

# =====================================
# 1. Ïû•Î•¥Î≥Ñ Î¨∏ÏÑú Ï§ÄÎπÑ (ÏòàÏãú Íµ¨Ï°∞)
# =====================================
# genre_tokens Îäî Ïù¥ÎØ∏ ÎßåÎì§Ïñ¥Ï†∏ ÏûàÎã§Í≥† Í∞ÄÏ†ï
# { genre : [token1, token2, ...] }

# ÏòàÏãú (ÎÑ§ ÏΩîÎìúÏóê Ïù¥ÎØ∏ ÏûàÏúºÎ©¥ Ïù¥ Î∏îÎ°ùÏùÄ Ï†úÍ±∞)
genre_tokens = {
    "Action": ["fight", "hero", "war", "fight", "gun"],
    "Romance": ["love", "heart", "kiss", "love"],
    "Comedy": ["funny", "joke", "laugh", "love"]
}

# =====================================
# 2. TF-IDF Î≤°ÌÑ∞ÎùºÏù¥Ï†Ä ÌïôÏäµ
# =====================================
documents = [
    " ".join(tokens)
    for tokens in genre_tokens.values()
]

vectorizer = TfidfVectorizer()
tfidf_matrix = vectorizer.fit_transform(documents)
feature_names = np.array(vectorizer.get_feature_names_out())

# =====================================
# 3. Ïû•Î•¥Î≥Ñ ÏÉÅÏúÑ TF-IDF ÌÇ§ÏõåÎìú Ï∂îÏ∂ú Ìï®Ïàò
# =====================================
def get_genre_keywords(genre, top_k=15):
    genre_idx = list(genre_tokens.keys()).index(genre)
    tfidf_scores = tfidf_matrix[genre_idx].toarray().flatten()

    top_indices = tfidf_scores.argsort()[::-1][:top_k]
    keywords = feature_names[top_indices]

    return keywords.tolist()

# =====================================
# 4. üîß genre_tfidf_words Ï†ïÏùò (ÏóêÎü¨ ÌïµÏã¨ Ìï¥Í≤∞)
# =====================================
genre_tfidf_words = {
    genre: set(get_genre_keywords(genre, top_k=15))
    for genre in genre_tokens.keys()
}

# =====================================
# 5. Í≥µÌÜµ Îã®Ïñ¥ / Ïû•Î•¥Î≥Ñ Í≥†Ïú† Îã®Ïñ¥
# =====================================
common_words = set.intersection(*genre_tfidf_words.values())

genre_unique_words = {
    genre: words - common_words
    for genre, words in genre_tfidf_words.items()
}

# =====================================
# 6. Í≤∞Í≥º Ï∂úÎ†•
# =====================================
print("üìå Í≥µÌÜµ Îã®Ïñ¥")
print(common_words)

print("\nüìå Ïû•Î•¥Î≥Ñ Í≥†Ïú† Îã®Ïñ¥")
for genre, words in genre_unique_words.items():
    print(f"{genre}: {sorted(words)}")
```

### ÏùºÎ∞òÎ™ÖÏÇ¨ + OOV Ï†úÍ±∞

```python
stopwords = {'ÏÇ¨Îûå', 'Ïù¥ÏïºÍ∏∞', 'ÏÇ¨Í±¥', 'Í¥ÄÍ≥Ñ', 'ÏãúÍ∞Ñ'}

genre_unique_words = {
    genre: {
        w for w in words
        if w not in stopwords and w in w2v_model.wv
    }
    for genre, words in genre_unique_words.items()
}
```

### Embedding Í∏∞Î∞ò ÏùòÎØ∏

```python
from sklearn.metrics.pairwise import cosine_similarity

def semantic_refine(words, model, top_k=8):
    words = list(words)
    if len(words) == 0:
        return []

    vectors = np.array([model.wv[w] for w in words])
    centroid = vectors.mean(axis=0).reshape(1, -1)
    sims = cosine_similarity(vectors, centroid).ravel()
    idx = sims.argsort()[::-1][:top_k]
    return [words[i] for i in idx]

genre_attributes = {
    genre: semantic_refine(words, w2v_model)
    for genre, words in genre_unique_words.items()
}
```

### WEAT Score

```python
import numpy as np

def cosine_similarity(a, b):
    return np.dot(a, b) / (np.linalg.norm(a) * np.linalg.norm(b))
```

```python
def association(w, A, B):
    return (
        np.mean([cosine_similarity(w2v_model.wv[w], w2v_model.wv[a]) for a in A]) -
        np.mean([cosine_similarity(w2v_model.wv[w], w2v_model.wv[b]) for b in B])
    )
```

```python
def weat_score(X, Y, A, B):
    return (
        np.mean([association(x, A, B) for x in X]) -
        np.mean([association(y, A, B) for y in Y])
    )
```

```python
weat_results = {}

for genre, words in genre_keywords.items():
    if len(words) < 5:
        continue

    score = weat_score(
        X=words,
        Y=commercial_words,
        A=art_words,
        B=commercial_words
    )
    weat_results[genre] = score
```
